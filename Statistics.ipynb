```python
import numpy as np 
import seaborn as sns
```

```python
def dist_func(x):
  return sns.histplot(data=x)
```

```python
#Uniform Distribution
x=np.random.uniform(0,100,1000)
dist_func(x)
```

```python
#T-Distribution
x=np.random.standard_t(50,1000)
dist_func(x)
```

```python
#Normal Distribution
x=np.random.normal(0,1,1000)
dist_func(x)
```

```python
#Binomial Distribution
x=np.random.binomial(size=1000, n=1, p= 0.3)
dist_func(x)
```

```python
#Poisson Distribution
x=np.random.poisson(5,1000)
dist_func(x)
```

```python
#Exponential Distribution
x=np.random.exponential(5,1000)
dist_func(x)
```

```python
#Beta Distribution
x=np.random.beta(1,1,1000)
dist_func(x)
```

```python
#Log-Normal Distributions
x=np.random.lognormal(0,1,10)
dist_func(x)
```
